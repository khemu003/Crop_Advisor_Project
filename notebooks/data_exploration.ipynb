```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Count images per class
data_dir = 'data/raw/New Plant Diseases Dataset(Augmented)/New Plant Diseases Dataset(Augmented)/train'
classes = os.listdir(data_dir)
counts = [len(os.listdir(os.path.join(data_dir, cls))) for cls in classes]

df = pd.DataFrame({'Class': classes, 'Count': counts})
df.plot(kind='bar', x='Class', y='Count', figsize=(12,6))
plt.title('Class Distribution')
plt.show()
```
